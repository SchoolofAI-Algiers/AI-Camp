```python
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

dataset = load_digits()
dataset.keys()
```

```python
# convert data to pd.DataFrame
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
dataset.target
```

```python
X = df
y = dataset.target
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X_scaled.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model_score = model.score(X_test, y_test)
model_score
```

```python
X
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.50)     # 0.9: variance between original and pca_data
X_pca = pca.fit_transform(X)
print(f"data shape {X.shape}, data shape with PCA: {X_pca.shape}")
```

```python
X_pca
```

```python
#pca.explained_variance_ratio_
pca.n_components_
```

```python
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model_pca = LogisticRegression(max_iter=10000)
model_pca.fit(X_train_pca, y_train)
pca_model_score = model_pca.score(X_test_pca, y_test)
pca_model_score
```

```python
print("Models score:")
print(f"without PCA: {model_score} \nwith PCA: {pca_model_score}")
```
