### **3. Matrix-Vector Multiplication**
It's basically applying a linear transformation on the vector, the transformation is defined by the column vectors composing the matrix. The first column vector of $A$ transforms the first component $u$, and the second vector of $A$ trasnforms the second component of $u$ : 

$$A.u = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} . \begin{bmatrix} 2 \\ 2 \end{bmatrix} = 2 \begin{bmatrix} 1  \\ 3 \end{bmatrix} + 2 \begin{bmatrix} 2 \\ 4 \end{bmatrix} = \begin{bmatrix} 2 \\ 6 \end{bmatrix} + \begin{bmatrix} 4 \\ 8 \end{bmatrix} = \begin{bmatrix} 6 \\ 14 \end{bmatrix}$$

Multiplyin a vector by a matrix is very similar to applying a function $f$ on a scalar $x$ to transform it into another scalar $y$ !
$$f(x) = y$$

# Calculus

Let's implement Gradient Descent from Scratch !

Let's consider the function : 
   $$f(x,y) = \frac{1}{2}x^2 + \frac{7}{2}y^2$$
   $$\nabla f(x,y) = \begin{bmatrix}
x  \\
7y \\
\end{bmatrix}$$

```python
def f(X):
    x = X[0]
    y = X[1]
    return (1/2)*(x**2) + (7/2)*(y**2)

def gradient_f(X):
    x = X[0]
    y = X[1]
    return np.array([x,7*y])
```

```python
plt.figure(figsize=(5,5)) 

N = 100 
x_values , y_values = np.linspace(-5,5,N) , np.linspace(-5,5,N)
x, y = np.meshgrid(x_values, y_values)
f_xy = f([x,y])

# Tracer f en 3D
ax = plt.axes(projection ='3d')
ax.plot_surface(x, y, f_xy, cmap=cm.RdPu) 

ax.set_xlabel('x', fontsize=18, fontweight='bold')
ax.set_ylabel('y', fontsize=18, fontweight='bold')
ax.set_zlabel('f(x,y)', labelpad=30, fontsize=18, fontweight='bold')
plt.show()
```

We want to **minimize** the function $f$.fWe can use Gradient Descent for that !

Gradient Descent works as follows :
$$x_{n} = x_{n-1} - \alpha\frac{\partial f}{\partial x}(x_{n-1})$$
$$y_{n} = y_{n-1} - \alpha\frac{\partial f}{\partial y}(y_{n-1})$$

But we can work directly with  vectors : 

$$X_{n} = X_{n-1} - \alpha\nabla{f{X_{n-1}}}$$
where : 
$$X_{n} = \begin{bmatrix} x_{n} \\ y_{n} \end{bmatrix}$$

```python
def gradient_descent(x0,lr,eps):
    f_history = [f(x0)] 
    x_history = [x0] 
    
    xn = x0 - lr*gradient_f(x0)
    xo = x0        # x_(n-1)
    nb = 0         # nombre d'iterations
    while (np.linalg.norm(xn-xo) > eps):
        xo = xn
        xn = xn - lr*gradient_f(xn)
        nb = nb + 1  
        
        # save 
        fx =  f(xn)
        f_history.append(fx)
        x_history.append(xn)
        
    return xn, nb, f_history, x_history
```

```python
first_point = [3,4.8]
learning_rate = 1/7
error_permission = 0.001

sol, nb, f_fix, x_fix = gradient_descent(first_point , learning_rate , error_permission)
print("Solution :")
print(f"   X*= ({sol[0]},{sol[1]})")
print(f"   f(X*)= {f(sol)}")
print("")
print(f"Number of iterations : {nb}")
```

```python
plt.figure(figsize=(10, 8))
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.title("Gradient Descent")
 
levels = np.arange(-1000,1000,20)


N = 100 
x_values , y_values = np.linspace( -10 , 10 , N) , np.linspace( -10 , 10 , N)
X_plot, Y_plot = np.meshgrid(x_values, y_values)


contours = plt.contour ( X_plot , Y_plot , f([X_plot,Y_plot]) , 30,levels=levels)
plt.clabel(contours)


x_fix = np.array(x_fix)
evolution_X_fix = x_fix[:, 0]
evolution_Y_fix = x_fix[:, 1]
plt.plot(evolution_X_fix, evolution_Y_fix,c="red")
plt.plot(evolution_X_fix, evolution_Y_fix, '.')

plt.show()
```

### Visualizing Quantiles and finding outliers

```python
# Create a boxplot using seaborn
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")  

# Use the boxplot function from seaborn
sns.boxplot(x='food_category', y='co2_emission', data=food_consumption ,showmeans=True)
#sns.stripplot(x = "food_category",y = "co2_emission",color = 'black',alpha = 0.3,data = df)

plt.xlabel('Food Category')
plt.ylabel('CO2 Emission')
plt.title('Boxplot of CO2 Emission by Food Category')
plt.show()
```

```python
# Finding an outlier
condition = (food_consumption['food_category'] == 'beef') & (food_consumption['co2_emission'] > 1500)
result = food_consumption[condition]['country'].unique()

print("Country with CO2 consumption of beef > 1500:", result)
```

```python
# Calculate the quartiles of co2_emission
print(np.quantile(food_consumption['co2_emission'], [0, 0.25, 0.5, 0.75, 1]))

# Calculate the quintiles of co2_emission
print(np.quantile(food_consumption['co2_emission'], [0, 0.2, 0.4, 0.6, 0.8, 1]))
```

```python
emissions_by_country = food_consumption.groupby('country')['co2_emission'].sum()

ax = emissions_by_country.plot(kind='bar', figsize=(26, 6))
mean_emission = emissions_by_country.mean()
median_emission = emissions_by_country.median()

ax.axhline(mean_emission, color='red', linestyle='dashed', linewidth=2, label='Mean Emission')
ax.axhline(median_emission, color='green', linestyle='dashed', linewidth=2, label='Median Emission')

plt.title('Total CO2 Emissions by Country')
plt.xlabel('Country')
plt.ylabel('Total CO2 Emissions')
plt.legend()
plt.show()
```
